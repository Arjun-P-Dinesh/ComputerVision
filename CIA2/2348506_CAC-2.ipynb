```python
import numpy as np
from scipy import ndimage as ndi
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import cv2
```

Question 1: <br>
1. Illustrate the following geometric operations using Affine transformation <br>
i) Translation<br>
ii) Scaling<br>
iii) Rotation<br>

```python
img = plt.imread("/home/arj/TriSem3/ComputerVision/CIA2/porsche_cgt.jpg")
```

```python
plt.imshow(img)
```

```python
img_g = rgb2gray(img)
```

```python
w, h = img_g.shape
w,h
```

```python
plt.imshow(img_g)
```

```python
#translation
tx,ty=50,400
mat_translate = np.array([[1,0,tx],[0,1,ty],[0,0,1]])
img_trns = ndi.affine_transform(img_g, mat_translate)
```

```python
#output of image applied with Translation Transformation
plt.imshow(img_trns)
```

```python
#scaling
s_x, s_y = 0.6, 0.5
mat_scale = np.array([[s_x,0,0],[0,s_y,0],[0,0,1]])
img_scale = ndi.affine_transform(img_g, mat_scale)
```

```python
#output of image applied with Scaling Transformation
plt.imshow(img_scale)
```

```python
#rotation
theta = np.pi/6
mat_rotate = np.array([[1,0,w/2],[0,1,h/2],[0,0,1]]) @ np.array([[np.cos(theta),np.sin(theta),0],[np.sin(theta),-np.cos(theta),0],[0,0,1]]) @ np.array([[1,0,- w/2],[0,1,-h/2],[0,0,1]])
img_rotate = ndi.affine_transform(img_g, mat_rotate)
```

```python
#output of image applied with Rotation Transformation
plt.imshow(img_rotate)
```

Question 2 :<br>
Implement the following Linear Spatial Filtering techniques<br>
i) Average filter<br>
ii) Laplacian Filter

```python
#original image 
plt.imshow(img)
```

```python
#Average Filter
kernel_size = (5, 5)
img_avg = cv2.blur(img, kernel_size)
```

```python
#output of image applied with Average filter
plt.imshow(img_avg)
```

```python
#Laplacian Filter
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1,  9, -1],
                              [-1, -1, -1]])
img_sharp = cv2.filter2D(img, -1, sharpening_kernel)
```

```python
#output of image applied with Laplacian Filter
plt.imshow(img_sharp)
```
