```python
!curl -o img.jpg https://buffer.com/cdn-cgi/image/w=1000,fit=contain,q=90,f=auto/library/content/images/size/w600/2023/10/free-images.jpg
import cv2
img = cv2.imread('img.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
# Convert to different formats using OpenCV
cv2.imwrite("output_image.jpg", img)  # Save as JPEG
cv2.imwrite("output_image.png", img)  # Save as PNG
cv2.imwrite("output_image.tiff", img)  # Save as TIFF
```

```python
#Using ImageIO
import imageio as iio
img = iio.imread("img.jpg")
```

```python
# write it in a new format
iio.imwrite("g4g.png", img)
```

```python
# importing matplotlib modules
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
# Read Images
img = mpimg.imread('img.jpg')
# Output Images
plt.imshow(img)
```

```python
# importing PIL
#Python Imaging Library
from PIL import Image

# Read image
img = Image.open('img.jpg')

# Output Images
#img.show(img)
display(img)

# prints format of image
print(img.format)
```

```python
# Convert to different formats using Pillow
img.save("output_image.jpeg")  # Save as JPEG
img.save("output_image.png")  # Save as PNG
img.save("output_image.tiff")  # Save as TIFF
```
